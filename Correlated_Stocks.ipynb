## Generating a Dataframe of Open Prices from Training Set for All Tech Stocks

```python
# Extract and rename columns from each DataFrame
Date = APPL_Training_Set.iloc[:, 0].rename('Date')
APPL_Open_Price = APPL_Training_Set.iloc[:, 1].rename('APPL_Open_Price')
AMZN_Open_Price = AMZN_Training_Set.iloc[:, 1].rename('AMZN_Open_Price')
ASML_Open_Price = ASML_Training_Set.iloc[:, 1].rename('ASML_Open_Price')
EA_Open_Price = EA_Training_Set.iloc[:, 1].rename('EA_Open_Price')
GD_Open_Price = GD_Training_Set.iloc[:, 1].rename('GD_Open_Price')
GOOG_Open_Price = GOOG_Training_Set.iloc[:, 1].rename('GOOG_Open_Price')
INTC_Open_Price = INTC_Training_Set.iloc[:, 1].rename('INTC_Open_Price')
META_Open_Price = META_Training_Set.iloc[:, 1].rename('META_Open_Price')
MSFT_Open_Price = MSFT_Training_Set.iloc[:, 1].rename('MSFT_Open_Price')
NOK_Open_Price = NOK_Training_Set.iloc[:, 1].rename('NOK_Open_Price')
NTDOY_Open_Price = NTDOY_Training_Set.iloc[:, 1].rename('NTDOY_Open_Price')
NVDA_Open_Price = NVDA_Training_Set.iloc[:, 1].rename('NVDA_Open_Price')
TSLA_Open_Price = TSLA_Training_Set.iloc[:, 1].rename('TSLA_Open_Price')

# Create a new DataFrame with these columns
Tech_Stocks_Open_Price_Training = pd.DataFrame({
    'Date': Date,
    'APPL_Open_Price': APPL_Open_Price,
    'AMZN_Open_Price': AMZN_Open_Price,
    'ASML_Open_Price': ASML_Open_Price,
    'EA_Open_Price': EA_Open_Price,
    'GD_Open_Price': GD_Open_Price,
    'GOOG_Open_Price': GOOG_Open_Price,
    'INTC_Open_Price': INTC_Open_Price,
    'META_Open_Price': META_Open_Price,
    'MSFT_Open_Price': MSFT_Open_Price,
    'NOK_Open_Price': NOK_Open_Price,
    'NTDOY_Open_Price': NTDOY_Open_Price,
    'NVDA_Open_Price': NVDA_Open_Price,
    'TSLA_Open_Price': TSLA_Open_Price
})
```

```python
# Adjust the indices if necessary
Tech_Stocks_Open_Price_Training = Tech_Stocks_Open_Price_Training.reset_index(drop=True)

# Display the new DataFrame
print(Tech_Stocks_Open_Price_Training.head())
```

## Correlation Heatmaps

```python
# Select specific columns (replace with your column names)
training_stock_open_prices_to_analyze = ['APPL_Open_Price', 'AMZN_Open_Price', 'ASML_Open_Price', 'EA_Open_Price', 'GD_Open_Price', 'GOOG_Open_Price', 'INTC_Open_Price', 'META_Open_Price', 'MSFT_Open_Price', 'NOK_Open_Price', 'NTDOY_Open_Price', 'NVDA_Open_Price', 'TSLA_Open_Price']

# Calculate the correlation matrix
corr_matrix = Tech_Stocks_Open_Price_Training[training_stock_open_prices_to_analyze].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
